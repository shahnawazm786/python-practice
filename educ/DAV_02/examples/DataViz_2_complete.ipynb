```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv('vgsales.csv')
data.head()
```

Boxplot

```python
plt.figure(figsize=(8,6))
sns.boxplot(data["Global_Sales"])
plt.yticks(fontsize=12)
plt.ylabel('Global Sales (in million dollars)', fontsize=12)
plt.title('Global Sales of video games', fontsize=12)
plt.show()
```

Continous vs Continous

```python
game_life = data.groupby(by = "Name").agg(min_year = ("Year", "min"),
                              max_year = ("Year", "max"))

game_life["range"] = game_life["max_year"] - game_life["min_year"]

game_life_sorted = game_life.sort_values(by = "range", ascending = False).head()
game_life_sorted.reset_index(inplace=True)

game_life_sorted
```

```python
game_life_sorted["Name"][0]
```

```python
ih=data[data["Name"]==game_life_sorted["Name"][0]]
```

```python
ih
```

```python
ih[["Year","NA_Sales"]]
```

```python
plt.title("Ice Hockey Sales Trend")
sns.lineplot(x="Year",y="NA_Sales",data=ih,color="r")
plt.show()
```

```python
def plot_trend(n):
  ih=data[data["Name"]==game_life_sorted["Name"][n]]
  plt.title(f'{game_life_sorted["Name"][n]}Sales Trend')
  sns.lineplot(x="Year",y="NA_Sales",data=ih,color="r")
  plt.show()
for i in range(5):
  plot_trend(n=i)
```

```python
ih=data[data["Name"]==game_life_sorted["Name"][0]]
bb=data[data["Name"]==game_life_sorted["Name"][1]]
sns.lineplot(x='Year',y='NA_Sales',data=ih,label="Ice Hockey")
sns.lineplot(x='Year',y='NA_Sales',data=bb,label="Baseball")
plt.legend(loc="upper right",ncol=2,frameon=False)
plt.show()
```

```python
ih=data[data["Name"]==game_life_sorted["Name"][0]]
bb=data[data["Name"]==game_life_sorted["Name"][1]]
sns.lineplot(x='Year',y='NA_Sales',data=[ih,bb],label=["Ice Hockey","Baseball"])
```

Dodged Countplot/Side-by-side barchart--> when you are comparing categorical vs categorical

```python
top3_pub=data["Publisher"].value_counts().index[:3]
top3_gen=data["Genre"].value_counts().index[:3]
top3_plat=data["Platform"].value_counts().index[:3]
top3_data=data.loc[(data["Publisher"].isin(top3_pub))& (data["Genre"].isin(top3_gen))& (data["Platform"].isin(top3_plat))]
top3_data.shape

```

```python
plt.figure(figsize=(6,4))
sns.countplot(x='Publisher',data=top3_data)
plt.ylabel("Count of Games")
```

Stacked Barchart

```python
df_stacked_plot=pd.crosstab(index=top3_data["Publisher"],columns=top3_data["Genre"])
```
