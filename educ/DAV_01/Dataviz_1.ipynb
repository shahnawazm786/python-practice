```python
data={
     "A":10,
     "B":20,
     "C":15



}
```

```python
import matplotlib.pyplot as plt
plt.bar(data.keys(),data.values())
plt.xlabel("Category")
plt.ylabel("Value")
plt.title("Bar Chart")
plt.show()
```

### How can we find top 5 genres?

```python
cat_counts=data["Genre"].value_counts()
cat_counts
```

```python
x_bar=cat_counts.index
y_bar=cat_counts
plt.bar(x_bar,y_bar)
```

```python
plt.figure(figsize=(12,5))
plt.bar(x_bar,y_bar,width=0.5,color="hotpink")
plt.xticks(rotation=45,fontsize=12)
plt.xlabel("Genre",fontsize=12)
plt.ylabel("Count",fontsize=12)
plt.title("Top 5 Genres",fontsize=20)
```
