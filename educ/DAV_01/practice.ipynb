```python
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
dic={'A':10,'B':20,'C':30}
plt.bar(x=dic.keys(),height=dic.values(),color="#86AEDB")
plt.xlabel('Count')
plt.ylabel('Name')
plt.title('Bar graph')
plt.show()
```

```python
df=pd.read_csv('final_vg.csv')
print(df.head())
```

```python
# show the columns
df.columns
```

```python
# Top 5 genere 
df_top_five_genre=df['Genre'].value_counts().head(15)
df_top_five_genre
```

```python
x_bars=df_top_five_genre.index
y_bar=df_top_five_genre
plt.bar(x_bars,y_bar)
```

```python
plt.figure(figsize=(15,5))
plt.bar(x_bars,y_bar,width=0.5,color='hotpink')
plt.xticks(rotation=60, fontsize=12) #[15,30,45,60,75,90]
plt.xlabel('Gener',fontsize=12)
plt.ylabel('Count',fontsize=12)
plt.title('Genere chart',fontsize=20)
plt.show()
```
